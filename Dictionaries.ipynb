## Update/Add an element to the dictionary 

##### Dictionaries are mutable, we can add new itmes or change the value of existing items using assignement operator. If the key is already present the value gets updated otherwise a new value pair is generated.

```python
myDict={'name': 'Edy', 'Age': 26}
print(myDict)
myDict['Age']=27
print(myDict)
```

#### Time Complexity for this is **O(1)**
#### Space Complexity for this is **O(1)**

```python
myDict['Address']='London'
myDict
```

##### This is how addition works in dictionaries
##### Time Complexity: **O(1)**
##### Space Complexity: **amortized O(1)**
##### In most cases the Space Complexity is **O(1)**, it changes to **O(n)** when data strcture needs to grow or shrink. That is why space complexity is amortised

## Traversing through a Dictionary

```python
myDict
```

```python
def traverseDict(Dict):
    for key in Dict: #----------------------->O(n)
        print(key, Dict[key]) #-------------->O(1)
```

```python
Dict=myDict
traverseDict(Dict)
```

#### Time Complexity : **O(n)**
#### Space Complexity: **O(1)**

## Searching an element in Dictionary

```python
def searchDict(Dict,value):
    for key in Dict: #---------------------------------------------------------------------->O(n)
        if Dict[key]==value: #-------------------------------------------------------------->O(1)
            return "Element found at " + str(key) + " and value is " + str(value) #--------->O(1)
    return "Element not found in Dictionary" #---------------------------------------------->O(1)
```

```python
myDict
```

```python
Dict=myDict
value=27
searchDict(Dict,value)
```

#### Time Complexity : **O(n)**
#### Space Complexity: **O(1)**

## Operations and Built in Functions

1. **in** operator
##### This operator determines whether an element exists in a collection or not in a dictionary
##### in operator only takes keys

```python
myDict={"one":"uno", "two":"dos", "three":"tres","four":"cuarto"}
'one' in myDict
```

```python
print('uno' in myDict)
print('uno' in myDict.values())
```

##### The *in* operator uses different algorithm for list and dictionaries. For list it uses Linear Search Algorithm. As the list gets longer the search time gets longer.
##### For dictionaries Python uses an algorithm called Hash Table that has a remarkable property. The *in* operator takes about the same amount of time no matter how many items are there in dictionary.
#### Time Complexity For *in* operator in Dictionary is **O(1)**

2. **for** operator
##### It's used for looping through any given collection and for dictionary as well.

```python
myDict={"one":"uno", "two":"dos", "three":"tres","four":"cuarto"}
for key in myDict:
    print(key)
```

#### Time complexity : **O(n)**
##### It has to visit all elements of dictionary one by one

### Python Build in fuction that work with Dictionaries

1. **all()**
##### It returns True when all elements in the given iterable are True, if not it returns False
##### syntax: all(dictionary)

| Cases       | Return Value|
| ----------- | ----------- |
| All Values are True      | True       |
| All Values are False      | False       |
| One Value is True (others are false)      | False       |
| One Value is False (others are True)     | False       |
| Empty Iterable      | True       |

```python
myDict1={1: True, 2:True}
print(all(myDict1))
myDict2={1:'True', 2:'True'}
print(all(myDict2))
myDict3={0:False, 1:False}
print(all(myDict3))
myDict4={0:True, 1:False, 2:False}
print(all(myDict4))
newDict={}
print(all(newDict))
```

2. **any()**
##### *any()* method returns True, if any element of collection is True, if not *any()* method will return False
##### syntax: any(dictionary)

| Cases       | Return Value|
| ----------- | ----------- |
| All Values are True      | True       |
| All Values are False      | False       |
| One Value is True (others are false)      | True      |
| One Value is False (others are True)     | True       |
| Empty Iterable      | False       |

```python
myDict1={1: True, 2:True}
print(any(myDict1))
myDict2={1:'True', 2:'True'}
print(any(myDict2))
myDict3={0:False, 1:False}
print(any(myDict3))
myDict4={0:True, 1:False, 2:False}
print(any(myDict4))
newDict={}
print(any(newDict))
```

3. **len()**
##### The *len()* function returns the number of items in the collection.
##### syntax: len(dictionary)
##### In case of dictionary len method returns the number of pairs

```python
myDict1={1: True, 2:True}
print(len(myDict1))
myDict2={1:'True', 2:'True'}
print(len(myDict2))
myDict3={0:False, 1:False}
print(len(myDict3))
myDict4={0:True, 1:False, 2:False}
print(len(myDict4))
newDict={}
print(len(newDict))
```

4. **sorted()**
##### This function sorts the elements of given collection in a specific order
##### syntax: sorted(iterable,reverse,key)
##### iterable: Any collection- list, tuple, string or dictionary
##### reverse: This is boolean parameter. If True , the sorted list is reversed (or sorted in descending order). Defaults to False if not provided.
##### key: This key function transforms each element before sorting, it takes the value and returns 1 value which is then used within sort instead of the original value

```python
myDict={'e':1, 'a':2, 'u':3, 'o':4, 'i':5}
print(sorted(myDict))
print(sorted(myDict,reverse=True))
```

```python
myDict={'ea':1, 'aas':2, 'udd':3, 'sseo':4, 'werwi':5}
print(sorted(myDict,key=len))
```

## Difference Between Dictionary and List

| Dictionary       | List|
| ----------- | ----------- |
| Unordered      | Ordered       |
| Access via keys      | Access via index       |
| Collection of key value pairs      | Collection of elements      |
| Preferred when you have unique key values     | Preferred when you have ordered data       |
| No duplicate members     | Allow  duplicate members       |

## Time and Space Complexity in Python Dictionary

| Cases       | Return Value | Space Complexity |
| ----------- | ----------- | ----------- |
| Creating a Dictionary      | O(len(dict))      | O(n) |
| Inserting a value in a Dictionary      | O(1)/O(n)       | O(1) |
| Traversing a given Dictionary      | O(n)      | O(1) |
| Accessing a given cell     | O(1)       | O(1) |
| Searching a given value      | O(n)       | O(1) |
| Deleting a given value      | O(1)       | O(1) |
