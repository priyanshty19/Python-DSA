## **Question-1**

### What is the runtime of the below code?

```python
def foo(array):
    sum=0 #1--------------------------------------------------->O(1)
    product=1 #2----------------------------------------------->O(1)
    
    for i in array: #3----------------------------------------->O(n)
        sum+=i #4---------------------------------------------->O(1)
        
    for i in array: #5----------------------------------------->O(n)
        product+=i #6------------------------------------------>O(1)
    print("Sum =" + str(sum) + ", product = "+ str(product))#7->O(1)
```

```python
array=[1,3,4,2,5]
foo(array)
```

##### From the rules of measuring time complexity it can be found out that the **Time Complexity of the following code = O(N)**

## **Question-3**

### What is the runtime of the below code?

```python
def printUnorderedPairs(array):
    for i in range(0,len(array)):
        for j in range(i+1, len(array)):
            print(str(array[i]) + "," + str(array[j]))
```

```python
array=[1,2,3,4,5]
printUnorderedPairs(array)
```

##### Another way of finding the runtime is counting iterations
###### 1st -> n-1
###### 2nd -> n-2
###### .
###### .
###### 1
###### (n-1) + (n-2) + (n-3) + ....+ 2 + 1
###### =1+2+3+.....+(n-3)+(n-2)+(n-1)
###### =n(n-1)/2
###### n<sup>2</sup>/2+n
##### Therefore **Time Complexity of the following code = O(N<sup>2</sup>)**

## **Question-6**

### What is the runtime of the below code?

```python
def reverse(array):
    for i in range(0,int(len(array)/2)): #1------>O(n)
        other = len(array)-i-1 #2---------------->O(1)
        temp = array[i] #3----------------------->O(1)
        array[i] = array[other] #4--------------->O(1)
        array[other] = temp #5------------------->O(1)
    print(array) #6------------------------------>O(1)
```

```python
array=[1,2,3,4,5]
reverse(array)
```

##### From the rules of measuring time complexity it can be found out that the **Time Complexity of the following code = O(N)**

## **Question-8**

### What is the runtime of below code?

```python
def factorial(n): #1-------------------------------->M(n)
    if n < 0: 
        return -1
    elif n == 0:
        return 1 #2------Till this point code take-->O(1)
    else:
        return n * factorial(n-1) #3---------------->M(n-1)

print(factorial(3))
```

## **Question-9**

### What is the runtime of the below code?

```python
def allFib(n):
    for i in range(n):
        print(str(i)+":, " + str(fib(i)))

def fib(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    return fib(n-1) + fib(n-2)
```

```python
n=5
allFib(n)
```

#### When we have a recurive function that makes multiple calls, the runtime often looks like
### ***O(Branches<sup>depth</sup>)***

##### By this our second function takes O(2<sup>N</sup>)

##### For our first function value of n is changing in the code
##### fib(1) -> 2<sup>1</sup> steps
##### fib(1) -> 2<sup>2</sup> steps
##### fib(1) -> 2<sup>3</sup> steps
##### fib(1) -> 2<sup>4</sup> steps
##### ...
##### fib(n) -> 2<sup>n</sup> steps
##### Total amount of work is sum of power of 2.
##### Total work = 2<sup>1</sup> + 2<sup>2</sup> + 2<sup>3</sup> + 2<sup>4</sup> + 2<sup>5</sup> + .... + 2<sup>n</sup> 
##### = 2<sup>n+1</sup>-2

##### Therefore the **Time Complexity = O(2<sup>N</sup>)**

```python
def powersOf2(n):
    if n < 1:
        return 0
    elif n == 1:
        print(1)
        return 1 #1------Till this point--->O(1)
    else:
        prev = powersOf2(int(n/2))
        curr = prev*2 # From here --
        print(curr)
        return curr # To here ------>O(1)

powersOf2(50)
```
